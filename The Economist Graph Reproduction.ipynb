This notebook recreates the donut chart from [this piece](https://www.economist.com/blogs/graphicdetail/2012/07/daily-chart-0) 

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42

%matplotlib inline
```

```python
df = pd.read_csv("data/bank-profits.csv")
df['percentage'] = df['year-2011'] / 786
```

```python
df.dtypes
```

```python
labels = 'Asia Pacific', 'Middle East and Africa', 'Latin America', 'Central and Eastern Europe', 'Western Europe','North America'
sizes = [148.554, 33.012, 19.650, 14.148, 363.132, 208.290]
colors = ['#1aa4d9', '#72d0f5', '#138a83', '#71bcbf','#6191a7','#09526c']
explode = (0.0, 0, 0, 0, 0, 0) 


plt.pie(sizes, explode=explode, labels=labels, colors=colors, startangle=220)
#plt.legend(labels, loc="best")  

plt.axis('equal')

centre_circle = plt.Circle((0,0),0.5,color='white', fc='white',linewidth=0.25)
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.tight_layout()
plt.savefig("bank_profits_2007.pdf", transparent=True)
```

```python
df = pd.read_csv("data/bank-profits.csv")

df.head(10)
```

```python
df['growth'] = df['year-2011']-df['year-2007']
df
```

```python
fig, ax = plt.subplots()
ax.hlines(xmin=[148.554, 33.012, 19.650, 14.148, 363.132, 208.290 ], xmax=[410.718, 49.530, 49.530, 25.908, 48.006, 178.308], 
          y=["Asia Pacific", "Middle East and Africa", "Latin America", "Central and Eastern Europe", "Western Europe", "North America"])
plt.tight_layout()
plt.savefig("the_bar_one.pdf", transparent=True)
```

```python
# The smartest vidulization on this probably is this: 
df.plot(kind='bar', x='region', y='growth', width= (0.1), figsize=(5,8))
ax.set_title("growth")

plt.tight_layout()
plt.savefig("bar_economist.pdf", transparent=True)
```
